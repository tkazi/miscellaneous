```python
# burlington
#converting pdf to text

pdf_files = os.listdir('../techplans/')

len(pdf_files)

def convert_pdf_to_txt(path):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = file(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos=set()
    for page in PDFPage.get_pages(fp,pagenos,maxpages=maxpages,password=password,caching=caching,check_extractable=True):
        interpreter.process_page(page)
    fp.close()
    device.close()
    str = retstr.getvalue()
    retstr.close()
    return str
```

```python
# burlington
# preprecoessing 

pdf_b = convert_pdf_to_txt('/Users/Taha/Desktop/datascience/techplans/burlington.pdf') 
from sklearn.feature_extraction.text import CountVectorizer
text_b = CountVectorizer().build_tokenizer()(pdf_b)

results_b = []
for t in text_b:
    tt = t.replace('\xe2','').replace("'",'')
    results_b.append(tt)

#print results
```

```python
# burlington
# preprecoessing -- removing stop words and reference of teacher, student and such

remove = ['district','technology','teacher','teachers','staff','student','students','use']
texts_b = []
for r in results_b:
    if r.lower().replace(' ','') not in stopwords.words('english') and r.lower().replace(' ','') not in remove:
        texts_b.append(r.lower())
        
#print texts 
```

```python
# burlington
# frequency counts

counts_b  = nltk.FreqDist(texts_b)
vocab   = len(counts_b.keys())
words   = sum(counts_b.values())
lexdiv  = float(words) / float(vocab)

print("Corpus has %i types and %i tokens for a lexical diversity of %0.3f" % (vocab, words, lexdiv))
```

```python
# topic modeling script from scikitlearn website

from __future__ import print_function
from time import time

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups

def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
    
```
