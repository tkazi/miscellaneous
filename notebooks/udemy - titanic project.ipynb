```python
def male_female_child(passenger):
    age, sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[ ['Age','Sex' ] ].apply(male_female_child, axis = 1)
```

```python
sns.factorplot('Pclass', data = titanic_df, hue = 'person', kind = 'count')
```

***
What deck were the passengers on and how does that relate to their class?
***

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])

cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']

cabin_df = cabin_df.sort_values(by='Cabin',ascending=True)

sns.factorplot('Cabin', data = cabin_df, palette = 'winter_d', kind = 'count')
    
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']

sns.factorplot('Cabin', data = cabin_df, palette = 'summer_d', kind = 'count')
```

***
Who was alone and who was with family?
***

```python
titanic_df['alone'] = titanic_df['SibSp'] + titanic_df['Parch']
```

```python
titanic_df['alone'].loc[titanic_df['alone'] > 0] = 'With Family'
titanic_df['alone'].loc[titanic_df['alone'] == 0] = 'Alone'
```

```python
sns.factorplot('alone', data = titanic_df, kind='count', palette='Blues')
```

```python
sns.factorplot('alone', data = titanic_df, kind='count', palette='Blues', hue='Pclass')
```

```python
titanic_df['Age'][(titanic_df['alone'] == 'Alone') & (titanic_df['Pclass'] == 3)].hist()
```
