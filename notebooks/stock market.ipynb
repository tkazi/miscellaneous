Correlation of Tesla stock (TSLA) and General Motors (GM)
***


```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
from pandas.io.data import DataReader
from datetime import datetime
from __future__ import division
```

```python
tech_list = ['TSLA','GM']

end = datetime.now()
start = datetime(end.year - 2, end.month, end.day)
```

```python
for stock in tech_list:
    globals()[stock] = DataReader(stock, 'yahoo', start, end)
```

```python
ts = TSLA
gm = GM 
```

```python
ts['Adj Close'].plot( legend=True, figsize=(20,4) )
```

```python
gm['Adj Close'].plot( legend=True, figsize=(20,4) )
```

```python
#moving averages for both tesla and gm for 15, 30, 60 and 90 days

ma_day = [15,30,60,90]

for ma in ma_day:
    column_name = 'MA %s days' %(str(ma))
    ts[column_name] = pd.rolling_mean(ts['Adj Close'], ma)
    gm[column_name] = pd.rolling_mean(gm['Adj Close'], ma)
```

```python
ts[ ['Adj Close', 'MA 30 days', 'MA 60 days'] ].plot(subplots=False, figsize=(20,5), legend=True)
plt.title('Tesla Moving Average (MA)', loc='right')
```

```python
gm[ ['Adj Close', 'MA 30 days', 'MA 60 days'] ].plot(subplots=False, figsize=(20,5), legend=True)
plt.title('General Motors Moving Average (MA)', loc='right')
```

```python
#both tesla and gm adjusted price in one data frame, and then get the daily return

closing_price = DataReader(tech_list, 'yahoo', start, end)['Adj Close']

daily_ret = closing_price.pct_change()
```

```python
sns.jointplot('TSLA','GM', closing_price, kind='scatter', color='purple')
```

```python
sns.jointplot('TSLA', 'GM', daily_ret, kind='scatter', color='purple')
```

```python
returns_fig = sns.PairGrid(daily_ret.dropna())

returns_fig.map_upper(plt.scatter, color='purple')

returns_fig.map_lower(sns.kdeplot, cmap='cool_d')

returns_fig.map_diag(plt.hist, bins=30)
```

```python
sns.corrplot(daily_ret.dropna(), annot=True)
```

```python
sns.corrplot(closing_price, annot=True)
```
